## which way of plotting is the best? pyplot vs matplotlib OO method?)

* when plotting quickly use plypot
* when plotting something advenced use OO method (or maybe plotly/sbs?)

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
#pyplot method
over_50.plot(kind="scatter", x='age', y='chol', c='target')
```

```python
# OO method mixed with pylot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c='target', ax=ax)

ax.set_xlim([45, 100])
```

```python
# OO from scrath

fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"] )

ax.set(title="Heart disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol");

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalac
fig, ax = plt.subplots(nrows=2, ncols = 1, figsize=(10,10), sharex=True)

scatter = ax[0].scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

ax[0].set(title="Cholesterol", ylabel="Cholesterol")
ax[0].legend(*scatter.legend_elements(), title="Target")
ax[0].axhline(y=over_50["chol"].mean(), linestyle='--')

scatter = ax[1].scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax[1].set(title="Thalach", xlabel="Age", ylabel="Thalach")
ax[1].legend(*scatter.legend_elements(), title="Target")
ax[1].axhline(y=over_50["thalach"].mean(), linestyle='--');

fig.suptitle("Heart disease analysis", fontsize= 16, fontweight="bold");
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter" )

ax.set(title="Heart disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol");

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(), linestyle='--');

```

```python
# subplot of chol, age, thalac
fig, ax = plt.subplots(nrows=2, ncols = 1, figsize=(10,10), sharex=True)

scatter = ax[0].scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")

ax[0].set(title="Cholesterol", ylabel="Cholesterol", ylim=(100,600))
ax[0].legend(*scatter.legend_elements(), title="Target")
ax[0].axhline(y=over_50["chol"].mean(), linestyle='--')

scatter = ax[1].scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")
ax[1].set(title="Thalach", xlabel="Age", ylabel="Thalach", xlim=(50,80), ylim=(60,200))
ax[1].legend(*scatter.legend_elements(), title="Target")
ax[1].axhline(y=over_50["thalach"].mean(), linestyle='--');

fig.suptitle("Heart disease analysis", fontsize= 16, fontweight="bold");
```
