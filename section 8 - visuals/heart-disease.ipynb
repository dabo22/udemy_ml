## which way of plotting is the best? pyplot vs matplotlib OO method?)

* when plotting quickly use plypot
* when plotting something advenced use OO method (or maybe plotly/sbs?)

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
#pyplot method
over_50.plot(kind="scatter", x='age', y='chol', c='target')
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c='target', ax=ax)

ax.set_xlim([45, 100])
```
