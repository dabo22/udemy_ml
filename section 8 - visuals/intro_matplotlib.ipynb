## Matplotlib example workflow

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [11,22,33,44]

fig, ax = plt.subplots(figsize=(8,4.5))

ax.plot(x,y)

ax.set(title="Simple Plt",
xlabel = "X-Axis",
ylabel = "Y-Axis")

fig.savefig("plot.png")
```

## Making figures with numpy arrays

We want:
* Line plot
* Scatter plot
* Bar Plot
* Histogram
* Subplots

```python
# create some data
x = np.linspace(0,10,100)
```

```python
# plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# use the same data for scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
n_b_p = {"Almond butter": 10,
        "Peanut butter": 8,
        "Cashew butter": 12}

fi, ax = plt.subplots()
ax.bar(n_b_p.keys(), n_b_p.values())
ax.set(title="Daniels nut Butter Store",
ylabel="Price (kr)");
```

```python
fig, ax = plt.subplots()

ax.barh(list(n_b_p.keys()), list(n_b_p.values()));
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(20,10))

ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(n_b_p.keys(), n_b_p.values());
ax4.hist(np.random.randn(1000));

fig.savefig("plot.png")
```

```python
# 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize=(10,5))

ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(n_b_p.keys(), n_b_p.values());
ax[1,1].hist(np.random.randn(1000));
```
